> ** Q. 지금까지 배운 반복문 조건문을 이용해서 놀이공원 입장료 계산 프로그램을 짜주세요**

- 입장료는 나이에 따릅니다
- 사용자로부터 입장할 사람의 수를 무한대로 받을 수 있습니다
- 사용자가 엔터를 입력하면 계산이 끝나고 지금까지 구한 입장료의 총합을 구해서 출력합니다

> ** 알고리즘 개발순서**
- 먼저 고객의 나이를 입력받아서 age라는 변수에 할당해준다
- 해당 변수(age) 의 값이 빈값이 아닌동안 계속 실행하는 반복문을 만들어 준다
- 조건문을 이용해서 해당 변수의 값이 연령대에 따라 total_price라는 변수에 합산될 수 있도록 분기처리한다.

```python
baby_price = 1000
child_price = 2000
adult_price = 3000
```

```python
baby_limit = 2
child_limit = 18
adult_limit = 60
```

```python
total_price = 0
age = input("고객의 나이를 입력하세요 (더이상 입력할 고객이 없으면 빈값에서 엔터를 누르세요)")

while age !='':
    age = int(age)
    if age <= baby_limit:
        total_price += baby_price
    elif age <= child_limit:
        total_price += child_price
    elif age <= adult_limit:
        total_price += adult_price
    age = input("고객의 나이를 입력하세요 (더이상 입력할 고객이 없으면 빈값에서 엔터를 누르세요)")
    
print ("놀이공원 입장료의 총액은 %d 입니다" %total_price)
```

```python
def sign(x):
    if x < 0:
        print("음수")
    else:
        print("양수")
    return sign
```

> **Q. 거래량이 전날보다 40%이상 갑자기 증가하면 signal을 주는 프로그램을 짜보자

- 결과물 : 전일 거래량(Volume)의 증감량
- 전일대비 증감율
- 증감율이 40%이상인 경우 'Over 40% UP' 아닌 경우 '이상없음'이라고 찍히게 하자

```python
samsung.Volume.diff()                      # 전후 증감 함수
```

> **Q. 증감량을 보여주는 컬럼을 추가해주세요

```python
samsung['증감량'] = samsung.Volume.diff()
```

```python
samsung
```

> **Q. 증감율을 보여주는 컬럼을 추가해주세요

```python
samsung['증감율'] = round((samsung.증감량 / (samsung.Volume - samsung.증감량)) * 100, 2)
```

```python
samsung
```

- 개발방법
- signal이라는 빈 리스트를 만들고
- if문과 for문을 써서 signal이라는 리스트에 up인지 이상없음 인지가 찍히게 함
- signal리스트를 컬럼에 할당

```python
signal = []                                                     # ★ 굉장히 중요한 기본적 코딩 ★★★★★★★★★
for row in samsung.증감율:
    if row >=40:
        signal.append('Over 40% UP')
    else:
        signal.append('이상없음')
```

```python
samsung['시그널'] = signal
```

```python
samsung
```

```python
df = pd.DataFrame(np.random.randn(10, 4))
df.iloc[:4, 1] = np.nan                                  # df.iloc[행, 열]      4번째행까지 1번째 열에 결측치를 채움
df.iloc[:2, 2] = np.nan                                  # df.iloc[행, 열]      2번째행까지 2번째 열에 결측치를 채움
df
```

```python
df.dropna(thresh=3)                                    #thresh 정상 데이터가   몇 개 이상이면 놔두고 지워(여기서는 3이면 2개이상)
```

```python
df.fillna(0)                                      # fillna  N/A값이 있으면 특정 값으로 채울것
```
