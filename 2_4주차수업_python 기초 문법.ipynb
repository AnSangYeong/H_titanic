> ** Q. 지금까지 배운 반복문 조건문을 이용해서 놀이공원 입장료 계산 프로그램을 짜주세요**
- 입장료는 나이에 따릅니다 
- 사용자로부터 입장할 사람의 수를 무한대로 받을 수 있습니다
- 사용자가 엔터를 입력하면 계산이 끝나고 지금까지 구한 입장료의 총합을 구해서 출력합니다

** 알고리즘 개발순서**
- 먼저 고객의 나이를 입력받아서 age라는 변수에 할당해준다
- 해당 변수(age) 의 값이 빈값이 아닌동안 계속 실행하는 반복문을 만들어 준다
- 조건문을 이용해서 해당 변수의 값이 연령대에 따라 total_price라는 변수에 합산될 수 있도록 분기처리한다.

```python
baby_price = 1000
child_price = 2000
adult_price = 3000
```

```python
baby_limit = 2
child_limit = 18
adult_limit = 60
```

```python
total_price = 0
age = input("고객의 나이를 입력하세요 (더이상 입력할 고객이 없으면 빈값에서 엔터를 누르세요)")

while age !='':
    age = int(age)
    if age <= baby_limit:
        total_price += baby_price
    elif age <= child_limit:
        total_price += child_price
    elif age <= adult_limit:
        total_price += adult_price

    age = input("고객의 나이를 입력하세요 (더이상 입력할 고객이 없으면 빈값에서 엔터를 누르세요)")
print ("놀이공원 입장료의 총액은 %d 입니다" %total_price)
```

> **Q. 사용자가 숫자를 입력하면 음수인지 양수인지를 판단하는 sign()이라는 함수를 만들어주세요 **

```python
def sign(x):
    if x < 0:
        print("음수")
    else:
        print("양수")
        
sign(1)
sign(-0.0000001)
```

```python
def sign(x):
    if x > 0:
        return "d%은 양수입니다"%x
    elif x < 0:
        return "d%은 음수입니다"%x
    else:
        return "0"

x = input('숫자를 입력하세요')
x = int(x)
sign(x)
```
